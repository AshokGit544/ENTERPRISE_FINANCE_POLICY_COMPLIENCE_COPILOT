```python
# ============================================================
# ENTERPRISE FINANCE POLICY COMPLIANCE COPILOT
# SAP FICO-STYLE CONTROLS + POLICY SEARCH + GROUNDED EXPLANATIONS
# RUN THIS ENTIRE CELL IN GOOGLE COLAB
# ============================================================

# =========================
# 1. INSTALL DEPENDENCIES
# =========================
!pip -q install pandas numpy matplotlib scikit-learn

# =========================
# 2. IMPORT LIBRARIES
# =========================
import warnings
warnings.filterwarnings("ignore")

import random
from datetime import datetime, timedelta
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

random.seed(42)
np.random.seed(42)

pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 30)

PROJECT_DIR = Path("enterprise_finance_policy_compliance_output")
PROJECT_DIR.mkdir(exist_ok=True)

print("Project directory:", PROJECT_DIR.resolve())

# =========================
# 3. MASTER DATA
# =========================
def make_vendors(n=85):
    vendor_bases = [
        "Alpha Grid Services", "Northwind Electric Supply", "GreenVolt Systems", "Prime Utility Parts",
        "TransEnergy Logistics", "Vertex Maintenance", "BluePeak Consulting", "SolarEdge Components",
        "Metro Industrial Tools", "Crescent Field Ops", "Pioneer Cable Works", "Helios Analytics",
        "Summit Procurement", "IronBridge Equipment", "EnerSys Contractors", "Utility Edge Networks"
    ]
    categories = ["Maintenance", "IT Services", "Electrical Components", "Logistics", "Consulting", "Field Services"]
    states = ["NY", "CT", "ME", "MA", "PA", "NJ", "RI"]
    rows = []
    for i in range(1, n + 1):
        rows.append({
            "vendor_id": f"V{i:04d}",
            "vendor_name": random.choice(vendor_bases) + f" {i}",
            "vendor_category": random.choice(categories),
            "vendor_state": random.choice(states),
            "risk_rating": random.choice(["Low", "Medium", "High"]),
            "payment_terms_days": random.choice([15, 30, 45, 60]),
            "preferred_vendor_flag": random.choice([1, 1, 1, 0])
        })
    return pd.DataFrame(rows)

def make_cost_centers():
    rows = [
        ("CC100", "Transmission Operations", "Operations"),
        ("CC110", "Distribution Maintenance", "Operations"),
        ("CC120", "Substation Engineering", "Engineering"),
        ("CC130", "Renewables Program", "Strategy"),
        ("CC140", "Customer Support", "Customer Ops"),
        ("CC150", "IT Shared Services", "Technology"),
        ("CC160", "Finance Operations", "Finance"),
        ("CC170", "Vegetation Management", "Field Ops"),
        ("CC180", "Grid Modernization", "Engineering"),
        ("CC190", "Regulatory Affairs", "Corporate")
    ]
    return pd.DataFrame(rows, columns=["cost_center_id", "cost_center_name", "department"])

def make_profit_centers():
    rows = [
        ("PC200", "Electric Distribution East", "East"),
        ("PC210", "Electric Distribution West", "West"),
        ("PC220", "Gas and Field Services", "North"),
        ("PC230", "Renewables and Storage", "Enterprise"),
        ("PC240", "Corporate Shared Services", "Corporate")
    ]
    return pd.DataFrame(rows, columns=["profit_center_id", "profit_center_name", "region_group"])

def make_gl_accounts():
    rows = [
        ("500100", "Opex - Maintenance", "Expense"),
        ("500200", "Opex - IT Services", "Expense"),
        ("500300", "Opex - Contractors", "Expense"),
        ("500400", "Opex - Logistics", "Expense"),
        ("500500", "Opex - Vegetation Mgmt", "Expense"),
        ("500600", "Opex - Consulting", "Expense"),
        ("500700", "Opex - Emergency Repairs", "Expense"),
        ("500800", "Capex - Grid Modernization", "Capital"),
        ("110100", "Accounts Payable", "Liability")
    ]
    return pd.DataFrame(rows, columns=["gl_account", "gl_account_name", "gl_type"])

vendors = make_vendors(90)
cost_centers = make_cost_centers()
profit_centers = make_profit_centers()
gl_accounts = make_gl_accounts()

print("Master data created:")
print("vendors:", vendors.shape)
print("cost_centers:", cost_centers.shape)
print("profit_centers:", profit_centers.shape)
print("gl_accounts:", gl_accounts.shape)

# =========================
# 4. POLICY / CONTROL LIBRARY
# =========================
policy_rules = [
    {
        "policy_id": "POL001",
        "policy_title": "Consulting Approval Threshold",
        "policy_area": "Approval",
        "policy_text": "Consulting expenses above 50000 USD require director approval and documented business justification before final posting."
    },
    {
        "policy_id": "POL002",
        "policy_title": "Emergency Repair Approval",
        "policy_area": "Emergency Spending",
        "policy_text": "Emergency repair expenses above 75000 USD require vice president approval and same-week compliance review."
    },
    {
        "policy_id": "POL003",
        "policy_title": "Non Preferred Vendor Control",
        "policy_area": "Vendor Governance",
        "policy_text": "Transactions with non preferred vendors above 25000 USD must include sourcing exception documentation and manager approval."
    },
    {
        "policy_id": "POL004",
        "policy_title": "Late Invoice Processing",
        "policy_area": "Accounts Payable",
        "policy_text": "Invoices submitted more than 45 days after posting require explanation and additional validation before payment release."
    },
    {
        "policy_id": "POL005",
        "policy_title": "High Risk Vendor Review",
        "policy_area": "Vendor Risk",
        "policy_text": "High risk vendors with expense transactions above 40000 USD must be reviewed by finance controls before payment."
    },
    {
        "policy_id": "POL006",
        "policy_title": "Vegetation Management Contractor Control",
        "policy_area": "Field Operations",
        "policy_text": "Vegetation management contractor spend above 60000 USD requires field operations lead approval and work package reference."
    },
    {
        "policy_id": "POL007",
        "policy_title": "IT Services Cap",
        "policy_area": "Technology",
        "policy_text": "IT services invoices above 55000 USD require cost center owner approval and proof of service completion."
    },
    {
        "policy_id": "POL008",
        "policy_title": "Weekend Posting Review",
        "policy_area": "Posting Controls",
        "policy_text": "Transactions posted on weekends above 30000 USD require review for urgency, authorization, and completeness."
    }
]

policy_df = pd.DataFrame(policy_rules)
policy_df.to_csv(PROJECT_DIR / "policy_library.csv", index=False)

print("\nPolicy library:")
display(policy_df)

# =========================
# 5. SYNTHETIC TRANSACTIONS
# =========================
descriptions = [
    "substation maintenance support",
    "vegetation clearing around power lines",
    "grid inspection contractor services",
    "cybersecurity monitoring support",
    "cloud cost optimization consulting",
    "field equipment transportation",
    "metering systems enhancement",
    "SCADA support services",
    "storm response contractor mobilization",
    "distribution pole inspection",
    "asset reliability assessment",
    "emergency transformer replacement",
    "temporary field crew support",
    "circuit restoration activities",
    "consulting support for vendor transition",
    "renewables program contractor support"
]

def generate_transactions(num_rows=4800):
    start_date = datetime(2024, 1, 1)
    end_date = datetime(2025, 12, 31)
    date_range_days = (end_date - start_date).days
    rows = []

    for i in range(1, num_rows + 1):
        vendor = vendors.sample(1).iloc[0]
        cc = cost_centers.sample(1).iloc[0]
        pc = profit_centers.sample(1).iloc[0]
        gl = gl_accounts.sample(1).iloc[0]

        posting_date = start_date + timedelta(days=random.randint(0, date_range_days))
        amount = round(random.uniform(3000, 85000), 2)

        if vendor["vendor_category"] == "Consulting":
            amount *= 1.20
        if cc["cost_center_name"] == "Vegetation Management":
            amount *= 1.15
        if gl["gl_account_name"] == "Opex - Emergency Repairs":
            amount *= 1.30
        if gl["gl_account_name"] == "Opex - IT Services":
            amount *= 1.10

        amount = round(amount, 2)

        approval_level = random.choice(["Manager", "Director", "VP"])
        justification_present = random.choice([1, 1, 1, 0])
        work_package_reference = random.choice([1, 1, 0])
        service_completion_proof = random.choice([1, 1, 1, 0])

        due_date = posting_date + timedelta(days=int(vendor["payment_terms_days"]))
        invoice_received_date = posting_date + timedelta(days=random.randint(0, 60))

        rows.append({
            "document_id": f"DOC{i:06d}",
            "invoice_id": f"INV{i:06d}",
            "posting_date": posting_date.date().isoformat(),
            "invoice_received_date": invoice_received_date.date().isoformat(),
            "due_date": due_date.date().isoformat(),
            "vendor_id": vendor["vendor_id"],
            "cost_center_id": cc["cost_center_id"],
            "profit_center_id": pc["profit_center_id"],
            "gl_account": gl["gl_account"],
            "amount_usd": amount,
            "currency": "USD",
            "description": random.choice(descriptions),
            "approval_level": approval_level,
            "justification_present": justification_present,
            "work_package_reference": work_package_reference,
            "service_completion_proof": service_completion_proof,
            "status": random.choice(["POSTED", "APPROVED", "PAID", "PENDING_REVIEW"])
        })

    return pd.DataFrame(rows)

finance_transactions = generate_transactions(4800)

# Inject extra policy-violation-style records
violation_idx = np.random.choice(finance_transactions.index, size=180, replace=False)

# consulting above threshold without director approval
consulting_idx = violation_idx[:45]
finance_transactions.loc[consulting_idx, "gl_account"] = "500600"
finance_transactions.loc[consulting_idx, "amount_usd"] = np.random.uniform(55000, 110000, size=len(consulting_idx)).round(2)
finance_transactions.loc[consulting_idx, "approval_level"] = "Manager"
finance_transactions.loc[consulting_idx, "justification_present"] = 0
finance_transactions.loc[consulting_idx, "description"] = "consulting support for vendor transition"

# emergency repair high value without VP approval
emergency_idx = violation_idx[45:90]
finance_transactions.loc[emergency_idx, "gl_account"] = "500700"
finance_transactions.loc[emergency_idx, "amount_usd"] = np.random.uniform(80000, 150000, size=len(emergency_idx)).round(2)
finance_transactions.loc[emergency_idx, "approval_level"] = "Director"
finance_transactions.loc[emergency_idx, "description"] = "emergency transformer replacement"

# non-preferred vendor over threshold
non_pref_idx = violation_idx[90:130]
non_pref_vendors = vendors[vendors["preferred_vendor_flag"] == 0]["vendor_id"].tolist()
if len(non_pref_vendors) == 0:
    non_pref_vendors = vendors["vendor_id"].sample(10, random_state=42).tolist()
finance_transactions.loc[non_pref_idx, "vendor_id"] = np.random.choice(non_pref_vendors, size=len(non_pref_idx), replace=True)
finance_transactions.loc[non_pref_idx, "amount_usd"] = np.random.uniform(26000, 70000, size=len(non_pref_idx)).round(2)
finance_transactions.loc[non_pref_idx, "justification_present"] = 0
finance_transactions.loc[non_pref_idx, "description"] = "field contractor sourcing exception"

# late invoice received
late_invoice_idx = violation_idx[130:155]
finance_transactions.loc[late_invoice_idx, "invoice_received_date"] = (
    pd.to_datetime(finance_transactions.loc[late_invoice_idx, "posting_date"]) + pd.to_timedelta(np.random.randint(46, 95, size=len(late_invoice_idx)), unit="D")
).dt.date.astype(str)
finance_transactions.loc[late_invoice_idx, "description"] = "late invoice submission requiring review"

# weekend posting large transactions
weekend_idx = violation_idx[155:]
weekend_dates = pd.to_datetime(["2025-01-04", "2025-01-05", "2025-02-08", "2025-02-09", "2025-03-01", "2025-03-02"])
finance_transactions.loc[weekend_idx, "posting_date"] = np.random.choice(weekend_dates.astype(str), size=len(weekend_idx))
finance_transactions.loc[weekend_idx, "amount_usd"] = np.random.uniform(32000, 90000, size=len(weekend_idx)).round(2)
finance_transactions.loc[weekend_idx, "description"] = "weekend posting review candidate"

finance_transactions.to_csv(PROJECT_DIR / "finance_transactions.csv", index=False)

print("\nTransactions created:", finance_transactions.shape)
display(finance_transactions.head())

# =========================
# 6. DATA MODEL JOIN
# =========================
df = (
    finance_transactions
    .merge(vendors, on="vendor_id", how="left")
    .merge(cost_centers, on="cost_center_id", how="left")
    .merge(profit_centers, on="profit_center_id", how="left")
    .merge(gl_accounts, on="gl_account", how="left")
)

df["posting_date"] = pd.to_datetime(df["posting_date"])
df["invoice_received_date"] = pd.to_datetime(df["invoice_received_date"])
df["due_date"] = pd.to_datetime(df["due_date"])
df["invoice_lag_days"] = (df["invoice_received_date"] - df["posting_date"]).dt.days
df["posting_weekday"] = df["posting_date"].dt.day_name()
df["is_weekend_posting"] = df["posting_date"].dt.weekday.isin([5, 6]).astype(int)
df["month"] = df["posting_date"].dt.to_period("M").astype(str)

approval_rank = {"Manager": 1, "Director": 2, "VP": 3}
df["approval_rank"] = df["approval_level"].map(approval_rank)

print("\nJoined data model shape:", df.shape)
display(df.head())

# =========================
# 7. RULE ENGINE
# =========================
def evaluate_compliance(row):
    violations = []

    if row["gl_account_name"] == "Opex - Consulting" and row["amount_usd"] > 50000:
        if row["approval_rank"] < 2 or row["justification_present"] == 0:
            violations.append("Consulting above 50000 without director approval and justification")

    if row["gl_account_name"] == "Opex - Emergency Repairs" and row["amount_usd"] > 75000:
        if row["approval_rank"] < 3:
            violations.append("Emergency repair above 75000 without VP approval")

    if row["preferred_vendor_flag"] == 0 and row["amount_usd"] > 25000:
        if row["justification_present"] == 0:
            violations.append("Non preferred vendor above 25000 without sourcing exception justification")

    if row["invoice_lag_days"] > 45:
        violations.append("Invoice received more than 45 days after posting")

    if row["risk_rating"] == "High" and row["amount_usd"] > 40000:
        violations.append("High risk vendor above 40000 requires finance controls review")

    if row["cost_center_name"] == "Vegetation Management" and row["amount_usd"] > 60000:
        if row["work_package_reference"] == 0 or row["approval_rank"] < 2:
            violations.append("Vegetation management contractor spend above 60000 missing required approvals or work package reference")

    if row["gl_account_name"] == "Opex - IT Services" and row["amount_usd"] > 55000:
        if row["approval_rank"] < 2 or row["service_completion_proof"] == 0:
            violations.append("IT services invoice above 55000 missing approval or proof of service completion")

    if row["is_weekend_posting"] == 1 and row["amount_usd"] > 30000:
        violations.append("Weekend posting above 30000 requires posting control review")

    return violations

df["violation_list"] = df.apply(evaluate_compliance, axis=1)
df["violation_count"] = df["violation_list"].apply(len)
df["is_non_compliant"] = (df["violation_count"] > 0).astype(int)
df["violation_text"] = df["violation_list"].apply(lambda x: " | ".join(x) if x else "No policy breach detected")

# =========================
# 8. POLICY MATCHING VIA SEMANTIC SEARCH
# =========================
policy_vectorizer = TfidfVectorizer(stop_words="english", ngram_range=(1, 2))
policy_matrix = policy_vectorizer.fit_transform(policy_df["policy_text"])

def retrieve_matching_policies(query_text, top_k=3):
    q_vec = policy_vectorizer.transform([query_text])
    sims = cosine_similarity(q_vec, policy_matrix)[0]
    idx = np.argsort(sims)[::-1][:top_k]
    result = policy_df.iloc[idx].copy()
    result["similarity_score"] = sims[idx]
    return result.reset_index(drop=True)

def attach_top_policies(row, top_k=2):
    query_text = f"{row['description']} {row['gl_account_name']} {row['cost_center_name']} {row['violation_text']} {row['vendor_category']}"
    matches = retrieve_matching_policies(query_text, top_k=top_k)
    policy_ids = matches["policy_id"].tolist()
    policy_titles = matches["policy_title"].tolist()
    return pd.Series({
        "matched_policy_ids": ", ".join(policy_ids),
        "matched_policy_titles": " | ".join(policy_titles)
    })

policy_matches = df.apply(attach_top_policies, axis=1)
df = pd.concat([df, policy_matches], axis=1)

# =========================
# 9. GROUNDED EXPLANATION
# =========================
def generate_grounded_explanation(row):
    status = "NON-COMPLIANT" if row["is_non_compliant"] == 1 else "COMPLIANT"
    return (
        f"Transaction {row['document_id']} for vendor {row['vendor_name']} is classified as {status}. "
        f"Amount is {row['amount_usd']:.2f} USD under {row['gl_account_name']} in cost center {row['cost_center_name']}. "
        f"Approval level recorded is {row['approval_level']}. "
        f"Preferred vendor flag is {row['preferred_vendor_flag']}. "
        f"Invoice lag is {row['invoice_lag_days']} days. "
        f"Weekend posting flag is {row['is_weekend_posting']}. "
        f"Detected compliance notes: {row['violation_text']}. "
        f"Most relevant policy references: {row['matched_policy_titles']}."
    )

df["grounded_explanation"] = df.apply(generate_grounded_explanation, axis=1)

# =========================
# 10. OUTPUT TABLES
# =========================
non_compliant_cases = (
    df[df["is_non_compliant"] == 1]
      .sort_values(["violation_count", "amount_usd"], ascending=[False, False])
      .copy()
)

top_non_compliant_cases = non_compliant_cases.head(25)

compliance_summary = (
    df.groupby("is_non_compliant", as_index=False)
      .agg(
          transaction_count=("document_id", "count"),
          total_amount_usd=("amount_usd", "sum")
      )
)
compliance_summary["compliance_status"] = compliance_summary["is_non_compliant"].map({0: "Compliant", 1: "Non-Compliant"})

violation_type_summary = (
    non_compliant_cases.assign(primary_violation=non_compliant_cases["violation_list"].apply(lambda x: x[0] if len(x) > 0 else "Unknown"))
    .groupby("primary_violation", as_index=False)
    .agg(
        transaction_count=("document_id", "count"),
        total_amount_usd=("amount_usd", "sum")
    )
    .sort_values("transaction_count", ascending=False)
)

policy_hit_summary = (
    df.assign(primary_policy=df["matched_policy_ids"].str.split(", ").str[0])
      .groupby("primary_policy", as_index=False)
      .agg(
          transaction_count=("document_id", "count"),
          total_amount_usd=("amount_usd", "sum")
      )
      .sort_values("transaction_count", ascending=False)
)

compliance_cases = df[[
    "document_id", "invoice_id", "vendor_name", "vendor_category", "risk_rating",
    "cost_center_name", "profit_center_name", "gl_account_name", "amount_usd",
    "approval_level", "invoice_lag_days", "is_weekend_posting", "is_non_compliant",
    "violation_count", "violation_text", "matched_policy_ids", "matched_policy_titles",
    "grounded_explanation"
]].copy()

# =========================
# 11. SAVE FILES
# =========================
vendors.to_csv(PROJECT_DIR / "vendors.csv", index=False)
cost_centers.to_csv(PROJECT_DIR / "cost_centers.csv", index=False)
profit_centers.to_csv(PROJECT_DIR / "profit_centers.csv", index=False)
gl_accounts.to_csv(PROJECT_DIR / "gl_accounts.csv", index=False)
top_non_compliant_cases.to_csv(PROJECT_DIR / "top_non_compliant_cases.csv", index=False)
compliance_summary.to_csv(PROJECT_DIR / "compliance_summary.csv", index=False)
violation_type_summary.to_csv(PROJECT_DIR / "violation_type_summary.csv", index=False)
policy_hit_summary.to_csv(PROJECT_DIR / "policy_hit_summary.csv", index=False)
compliance_cases.to_csv(PROJECT_DIR / "compliance_cases.csv", index=False)

# =========================
# 12. VISUALIZATIONS
# =========================
top_violation_chart = violation_type_summary.head(10)

plt.figure(figsize=(12, 5))
plt.bar(top_violation_chart["primary_violation"], top_violation_chart["transaction_count"])
plt.xticks(rotation=45, ha="right")
plt.title("Top Policy Violation Types")
plt.ylabel("Transaction Count")
plt.tight_layout()
plt.show()

monthly_non_compliance = (
    df.groupby(["month", "is_non_compliant"], as_index=False)
      .agg(transaction_count=("document_id", "count"))
)
monthly_non_compliance = monthly_non_compliance[monthly_non_compliance["is_non_compliant"] == 1]

plt.figure(figsize=(12, 5))
plt.plot(monthly_non_compliance["month"], monthly_non_compliance["transaction_count"], marker="o")
plt.xticks(rotation=45, ha="right")
plt.title("Monthly Non-Compliant Transaction Trend")
plt.ylabel("Transaction Count")
plt.tight_layout()
plt.show()

# =========================
# 13. INTERACTIVE POLICY SEARCH
# =========================
def search_policies(user_query, top_k=5):
    matches = retrieve_matching_policies(user_query, top_k=top_k)
    return matches

sample_policy_query = "high value consulting without director approval and justification"
sample_policy_results = search_policies(sample_policy_query, top_k=5)

print("\nSample policy search results:")
display(sample_policy_results)

# =========================
# 14. INTERACTIVE COMPLIANCE CASE SEARCH
# =========================
case_vectorizer = TfidfVectorizer(stop_words="english", ngram_range=(1, 2), max_features=5000)
case_matrix = case_vectorizer.fit_transform(compliance_cases["grounded_explanation"])

def search_compliance_cases(user_query, top_k=5):
    q_vec = case_vectorizer.transform([user_query])
    sims = cosine_similarity(q_vec, case_matrix)[0]
    idx = np.argsort(sims)[::-1][:top_k]
    result = compliance_cases.iloc[idx].copy()
    result["similarity_score"] = sims[idx]
    return result.reset_index(drop=True)

sample_case_query = "non preferred vendor above threshold without justification"
sample_case_results = search_compliance_cases(sample_case_query, top_k=5)

print("\nSample compliance case search results:")
display(sample_case_results)

# =========================
# 15. SHOW OUTPUTS
# =========================
print("\nCompliance summary:")
display(compliance_summary)

print("\nTop non-compliant cases:")
display(top_non_compliant_cases[[
    "document_id", "vendor_name", "cost_center_name", "gl_account_name", "amount_usd",
    "approval_level", "violation_count", "matched_policy_ids", "violation_text"
]])

print("\nGenerated files:")
for p in sorted(PROJECT_DIR.glob("*")):
    print("-", p.name)
```
